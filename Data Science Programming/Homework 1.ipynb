### 1.    Rock, Paper, Scissors (20 points)
This is a popular game (see here), and you must create a computer player.

**Game setup.** Each game round consists two turns, the first by the com-puter and the second by a human.  The computer continues playing rounds until the human chooses to quit.
- The computer chooses one of Rock, Paper, and Scissors, but keeps its choice secret.
- The computer asks for the human’s input.
- The human chooses one of Rock, Paper, and Scissors, or Quit.
- Unless  the  human  quits,  the  computer  figures  out  the  result  of  the game, as follows:
    - Rock smashes Scissors, so Rock beats Scissors.
    - Scissors can cut up paper, so Scissors beat Paper.
    - Paper covers Rock, so Paper beats Rock.
If both players chose the same, it is a draw.  The computer reports the result of this round.1
- If the human chooses to quit, the computer reports:
    - the number of games played, and
    - the number of times the human won.

**Computer’s brains.** The computer must be able to exploit some humanbiases.   If  the  human  has  played  Rock  most  often,  the  computer  shouldassume  that  he  or  she  will  play  Rock  in  the  next  round,  so  the  computer should play Paper.  If the human has played Rock and Paper equally often,and Scissors less often, the computer should assume that the human is goingto play either Rock or Paper (both equally likely) in the next round.  (What should the computer play?)
Hence, your program should remember how many Rock, Paper, or  Scissors  were  played  by  the  human. Note  that  we  don’t  need  to remember the order in which the human chooses these; the total counts so far for each choice will be enough.
 
 **Gotchas.**
- User input: How  you  want  to  receive  the  user’s  input  is  up  to  you, but you must check the user’s input to make sure it is valid (you can assume that the user input is of the correct type).  If it isn’t, request the user for input again.

```python
#Question1
import random
choiceOptions= ("Rock","Paper","Scissors")
paper = 0
rock = 0
scissors = 0
rounds = 0 
winningRounds = 0
userInput = "play"
while userInput != "quit":
    userChoice = input("Pick Rock, Paper or Scissors or type quit to Exit: ")
    if userChoice == "quit":
        print("Thanks for playing")
        userInput = "quit"
        print("Number of Rounds played",rounds)
        print("Number of times you won",winningRounds)
        break
    elif userChoice != "Rock" and userChoice != "Paper" and userChoice != "Scissors":
        print("Please check your spelling and try again")
        continue
    else:
        rounds +=1


    if rock==0 and paper == 0 and scissors == 0:
        computerChoice = random.choice(choiceOptions)
    elif rock == paper and paper == scissors:
        computerChoice = random.choice(choiceOptions)
    elif rock > paper and rock > scissors:
        computerChoice = "Paper"
    elif paper > rock and paper > scissors:
        computerChoice = "Scissors"
    elif scissors > paper and scissors > rock:
        computerChoice = "Rock"


    if userChoice == "Rock":
        if rock==paper and rock>scissors:
            computerChoice = "Paper"
        elif scissors==rock and scissors>paper:
            computerChoice = "Rock"
        rock +=1
    elif userChoice == "Paper":
        if paper==scissors and paper>rock:
            computerChoice = "Scissors"
        elif rock==paper and rock>scissors:
            computerChoice = "Paper"
        paper +=1
    elif userChoice == "Scissors":
        if paper==scissors and paper>rock:
            computerChoice = "Scissors"
        elif scissors==rock and scissors>paper:
            computerChoice = "Rock"
        scissors +=1

    if computerChoice == userChoice:
        print("The computer choice is",computerChoice)
        print("It is a tie, please try again")
        winningRounds +=0
        userInput = "1"
    elif computerChoice == "Rock":
        print("The computer choice is",computerChoice)
        if userChoice == "Paper":
            print("Paper beats Rock, you win!")
            winningRounds +=1
            userInput = "1"
        else:
            print("Rock beats Scissors, the computer wins!!")
            userInput = "1"
    elif computerChoice == "Scissors":
        print("The computer choice is",computerChoice)
        if userChoice == "Rock":
            print("Rock beats Scissors, you win!")
            winningRounds +=1
            userInput = "1"
        else:
            print("Scissors beats Paper, the computer wins!")
            userInput = "1"
    elif computerChoice == "Paper":
        print("The computer choice is",computerChoice)
        if userChoice == "Scissors":
            print("Scissors beats Paper, you win!")
            winningRounds +=1
            userInput = "1"
        else:
            print("Paper beats Rock, the computer wins!")
            userInput = "1"

```

```python
#Question2
import re

fp = open("FloridaVoters.html",'r') #open file for reading
voting_list = []           #initializing empty list to store data
for item in fp:            #iterating through the file and applying regex to separate relevant items by using html tags
    counties = re.findall('<td>', item)
    if len(counties) > 0:
        county = item.lstrip('<td>').rstrip('</td>\n') 
        voting_list.append(county)   #appending each item to a list

voting_table = []
for i in range(0,len(voting_list),3):           #grouping every thre elements together into voting_table list
    voting_table.append(voting_list[i:i+3])
# print(voting_table) 
county_table = []                               
for i in range(0,len(voting_table),2):          #add every other element from voting_table to county_table to remove "minor", "none", and "column" vote amounts
    county_table.append(voting_table[i])
for item in county_table:                       #removing commas from vote amounts and changing these values to integers
    for i in range(len(item)-1):
        item[i+1] = item[i+1].replace(',','')
        item[i+1] = int(item[i+1])


for i in range(len(county_table)):              #converting from a list of list to a list of tuples
    county_table[i] = tuple(county_table[i])


def democrat_voteCount(tuple_item):             #function to retrieve the third item (2nd index) of each tuple
        return tuple_item[2]


sorted_county_table = sorted(county_table, key=democrat_voteCount)  #creating a new sorted_country_table list for sorting based on democratic votes
for item in sorted_county_table:                                    #printing out each item of each tuple in the sorted_country_table list 
    for i in range(len(item)):
        print(item[i], end = ' ')
    print('\n')
```

### 3.    Near-duplicate detection (20 points)

Tweets  on  a  subject  are  often  nearly  (but  not  exactly)  duplicates  of  each other.   The  file Santa.txt contains  a  few  tweets  about  Santa  Claus,  one tweet per line.  We will try to detect tweets that we have already seen before.
- **(a)  Convert  each  tweet  into  a  dictionary  of  phrases  (6  points)**. Write a function called movingwindow that takes as input a tweet (that is,a string), and converts it into a bunch of phrases.  Each phrase is three con-secutive words in the tweet.  For example, the tweet‘‘This is an awesometweet, dude’’consists of the phrases‘‘ This is an’’,‘‘is an awesome’’,‘‘an awesome tweet,’’, and‘‘awesome tweet, dude’’.  Return a dictio-nary whose keys are these phrases, and just set the corresponding values to1.  This dictionary contains all the unique 3-word phrases in the tweet.
- **(b) Calculate similarity between two tweets (6 points)**. To check if one tweet is a near-duplicate of another, we compute their cosine similarity: cosine(tweet1, tweet2)=$\frac{matches}{\sqrt{n1∗n2}}$ , wherematches is  the  number  of  3-word  phrases  in  common  between  the two  tweets,  and n1 and n2 are  the  number  of  phrases  in  the  two  tweets respectively. Write a function called cosine that takes as input two dictionaries.  Each dictionary contains the 3-word phrases from one tweet.  Return the cosine similarity between the phrases in the two dictionaries.
Hint:You can use the functionmath.sqrt(x)to calculate the square-rootof any floating point numberx, but you must import the math module.
- **(c) Read in tweets, and output near-duplicates (8 points)**. Read in the tweets in the file Santa.txt.  For each tweet, figure out if it is a near-duplicate of any of the previously-seen tweets.  We say that the two tweets are near-duplicates if their cosine similarity is greater than 0.5. You should call the functions movingwindow and cosine here.

```python
#Question3
import math 
import re
    
def moving_window(string): 
    """Takes as input a tweet (that is a string), and converts it into a bunch of phrases.
    Each phrase is three consecutive words in the tweet.  For example, the tweet ‘‘This is an awesome tweet, dude’’
    consists of the phrases ‘‘This is an’’, ‘‘is an awesome’’, ‘‘an awesome tweet,’’, and ‘‘awesome tweet, dude’’.  
    Returns a dictionary whose keys are these phrases, and their corresponding values are 1.  
    This dictionary contains all the unique 3-word phrases in the tweet. """
    dict_phrase = {}
    items = string.split()
    for i in range(len(items)-2):
        group = items[i:i+3]
        phrase = ' '.join([(str(elem)).lower() for elem in group])
        if phrase not in dict_phrase:
            dict_phrase[phrase] = 1
    return dict_phrase

def cosine(dict1, dict2):
    """Takes as input two dictionaries.  
    Each dictionary contains the 3-word phrases from one tweet.  
    Returns the cosine similarity between the phrases in the two dictionaries."""
    matches = 0 
    for key in dict1:
        if key in dict2:
            matches += 1
    n1 = len(dict1)
    n2 = len(dict2)
    return matches/math.sqrt(n1*n2)

fp = open('Santa.txt', 'r')

list_of_tweets = []
all_tweets = []
line = fp.readline()
while line != '':
    line = line.rstrip('\n')
    all_tweets.append(line)
    line = fp.readline()

for i in range(len(all_tweets)):
    tweet_dict = moving_window(all_tweets[i])
    list_of_tweets.append(tweet_dict) 
#pint(tweet_dict)

#To see the cosine similarities between the tweets and find the near duplicate pairs
cosine_threshold = 0.5
near_duplicate_counter = 0
for x in range(len(list_of_tweets)):
    for y in range(len(list_of_tweets)):
        if x < y:
            cosine_result = cosine(list_of_tweets[x], list_of_tweets[y])
            if cosine_result > cosine_threshold:
                near_duplicate_counter += 1
                print("Tweet:",all_tweets[x],'\nTweet:', all_tweets[y], '\nThese tweets are near duplicates as they have a cosine similarity of ', cosine_result,'\n')
            #print("Tweet:",all_tweets[x],'\nTweet:', all_tweets[y], '\nThese tweets have a cosine similarity of ', cosine_result,'\n')    
print("There are", near_duplicate_counter, "instances of near duplicates among all these tweets")
```
